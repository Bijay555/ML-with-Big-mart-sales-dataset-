```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import preprocessing
  
def estimate_coef(x, y): 
    # number of observations/points 
    n = np.size(x) 
  
    # mean of x and y vector 
    m_x, m_y = np.mean(x), np.mean(y) 
  
    # calculating cross-deviation and deviation about x 
    SS_xy = np.sum(y*x) - n*m_y*m_x 
    SS_xx = np.sum(x*x) - n*m_x*m_x 
  
    # calculating regression coefficients 
    b_1 = SS_xy / SS_xx 
    b_0 = m_y - b_1*m_x 
  
    return(b_0, b_1) 
  
def plot_regression_line(x, y, b): 
    # plotting the actual points as scatter plot 
    plt.scatter(x, y, color = "m", 
               marker = "o", s = 30) 
  
    # predicted response vector 
    y_pred = b[0] + b[1]*x
    
    # plotting the regression line 
    plt.plot(x, y_pred, color = "g") 
  
    # putting labels 
    plt.xlabel('x') 
    plt.ylabel('y') 
  
    # function to show plot 
    plt.show()
    
def sum_square_error(y,y_pred):
    l=len(y)
    for i in range(l):
        sse=np.sum((y[i]-y_pred[i])**2)
    return sse

def total_sum_squared_error(y,m_y):
    l=len(y)
    for i in range(l):
        sst=np.sum((y[i]-m_y)**2)
    return sst

def r_squared(sse,sst):
    r=1-sse/sst
    return r

    
  
def main(): 
    # observations
     
    x = train.Item_MRP
    y = train.Item_Outlet_Sales
    #x_scaled = np.log(x)
    #y_scaled = np.log(y)
    x_a=x.values.reshape(-1,1)
    y_a=y.values.reshape(-1,1)
    std_scale = preprocessing.StandardScaler().fit(x_a)
    x_scaled = std_scale.transform(x_a)
    std_scale = preprocessing.StandardScaler().fit(y_a)
    y_scaled = std_scale.transform(y_a)
                                   
    y_pred=[]
    l=len(x)
    # estimating coefficients 
    b = estimate_coef(x_scaled, y_scaled) 
    print("Estimated coefficients:\nb_0 = {}   \nb_1 = {}".format(b[0], b[1]))
    n = np.size(x)
    for i in range(n):
        y_predict = b[0] + b[1]*x_scaled
        y_pred.append(y_predict)
    m_y=np.mean(y_scaled)
    
    # plotting regression line 
    plot_regression_line(x_scaled, y_scaled, b) 
    
    #predicting sum square error
    s=sum_square_error(y_scaled,y_pred)
    print("Sum square error:",s)
    
    #predicting total sum square error
    st=total_sum_squared_error(y_scaled, m_y)
    print("Total sum square error:",st)
    
    #prdicting r_squared value
    r = r_squared(s,st)
    print("R squared value:",r)
  
if __name__ == "__main__": 
    main() 
```

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
train=pd.read_csv("Train.csv")
from tensorflow import keras
```

```python
x = train.Item_MRP
y = train.Item_Outlet_Sales
x_a=x.values.reshape(-1,1)
y_a=y.values.reshape(-1,1)
std_scale = preprocessing.StandardScaler().fit(x_a)
    
x_scaled = std_scale.transform(x_a)
std_scale = preprocessing.StandardScaler().fit(y_a)
y_scaled = std_scale.transform(y_a)
    
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2) 
                                        
X_2 = poly.fit_transform(x_scaled.reshape(-1,1))
print(X_2.shape)
print(X_2[0])
print(X_2[1])
print(X_2[2])
```

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[3])])

optimizer = tf.keras.optimizers.Adam(0.3)

model.compile(optimizer=optimizer, loss='mean_squared_error')
tf_history = model.fit(X_2, y_scaled, epochs=50, verbose=True)

plt.plot(tf_history.history['loss'])
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.show()

mse = tf_history.history['loss'][-1]
y_hat = model.predict(X_2)

plt.figure(figsize=(12,7))
plt.title('TensorFlow Model')
plt.scatter(X_2[:, 1], y_scaled, label='Data $(X, y)$')
plt.plot(X_2[:, 1], y_hat, color='red', label='Predicted Line $y = f(X)$',linewidth=4.0)
plt.xlabel('$X$', fontsize=20)
plt.ylabel('$y$', fontsize=20)
plt.text(0,0.70,'MSE = {:.3f}'.format(mse), fontsize=20)
plt.grid(True)
plt.legend(fontsize=20)
plt.show()
```
